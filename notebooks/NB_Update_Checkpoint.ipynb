```python
from pyspark.sql.functions import current_timestamp

# Read copied files directly from OneLake
df = spark.read.format("binaryFile") \
    .load("Files/csv10to100/*")

result = (
    df.selectExpr(
        "path as file_path",
        "md5(path) as file_sha"
    )
    .withColumn("last_loaded_ts", current_timestamp())
)

result.write.mode("append").saveAsTable("dbo.file_checkpoint")

print("Checkpoint updated")
```
